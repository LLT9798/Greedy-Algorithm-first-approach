```python
import pandas as pd
import numpy as np

#import data and drop index
test = pd.read_csv('AssignmentProblem10by10.csv')
del test['Unnamed: 0']
test.head(10)
```

```python
#transfer data from to array
testarray = test.values
```

```python
testarray[0,2]
```

# The Goal: Minimize the cost of assignment
# Rule 1: Employees A-J ; Tasks 0 - 10
# Rule 2: Every employee can be assigned only one task and vice versa
# Rule 3: Use Greedy Algorithm to approach the goal


```python
pool1 = []
pool = []
while  True:
    for j in range(0,10):
        for k in range(0,10): #use two for loop to make sure go through every elements in the array
            if testarray[k,j] not in pool:  #this condition make sure the task not been assign to another employee
                pool1.append(testarray[k,j])
        pool.append(min(pool1))
        pool1 = []  #clean the list b/c we want to find the min cost in each column 
            
    break

print(pool)
```
